```python
# 載入檔名
source_path = r'./csv'
file_list = os.listdir(source_path)
file_list
```

```python
# 載入每個csv檔的內容
df_list = []
for article in file_list:
    info_list=[]
    try:
        with open(source_path + '/' + article, 'r', encoding='utf-8') as f:
            tem_str = f.read()
        info_list = tem_str.split('\n')
        info_list[0] = info_list[0].replace('\ufeff','')
        info_list = info_list[0:3]      
        df_list.append(info_list)
    except:
        print()
df_list
```

```python
# 建立dataframe
new_df = pd.DataFrame(df_list, columns=['作者','月份', '地點'])
new_df
```

```python
location_cnt_df = new_df.groupby(['地點'],as_index=False)['地點'].agg({'cnt':'count'})
location_cnt_df
```

```python
# 設定中文字型(可以從妳電腦裡找，改成妳喜歡的)，不然會亂碼
myfont = matplotlib.font_manager.FontProperties(fname='C:\Windows\Fonts\DFT_TX3.TTC')

# 這個function是為了呈現各縣市數值
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return ' ({v:d})'.format(v=val)
    return my_autopct

# 畫圓餅圖
patches,l_text,p_text = plt.pie(location_cnt_df['cnt'], labels=location_cnt_df['地點'], autopct=make_autopct(location_cnt_df['cnt']))
for t in l_text: 
    t.set_fontproperties(myfont)

# 圖例的位置
plt.legend(prop=myfont,loc=(1.2, 0.05))
plt.show()   
```
